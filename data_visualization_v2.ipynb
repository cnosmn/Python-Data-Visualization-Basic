```python
import seaborn as sb
dat1=sb.load_dataset("tips")
dat1.head(15)
```

```python
dat1.dtypes
```

```python
dat1.info()
```

```python
dat1.isnull().values.any()
```

```python
dat1.isnull().values.sum()
```

```python
dat1["sex"].value_counts().plot.barh().set_title("cinsiyet sayısı");
```

```python
dat1[["sex","smoker"]].value_counts().plot.barh().set_title("cinsiyet sayısı");
```

```python
sb.barplot(x="smoker",y=dat1.smoker.index,data=dat1)
#çubuk grafiği
```

```python
sb.kdeplot(dat1.total_bill,shade=True)
#yoğunluk grafiği
```

```python
sb.displot(dat1,x="tip",bins=50,kde=False)
#tek değişkenli dağılım hızlıca bakmanın yöntemi
```

```python
###   KORELASYON GRAFİKLERİ
sb.scatterplot(x="total_bill",y="tip",hue="sex",data=dat1)
#nokta grafiği
```

```python
sb.scatterplot(x="total_bill",y="tip",hue="sex",style="smoker",data=dat1)
```

```python
#doğrusal ilişki gösterimi

sb.lmplot(x="total_bill",y="tip",data=dat1,col="time");
```

```python
sb.barplot(x=dat1.time,y=dat1.time.index)
```

```python
sb.pairplot(dat1,hue="sex")
#verilerin kendi aralarında ilişkileri
```

```python
sb.lineplot(x="tip",y="size",hue="sex",data=dat1)
#:(
```
